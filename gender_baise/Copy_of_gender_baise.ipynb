```python
import pandas as pd
import numpy as np
import csv
from sklearn.linear_model import LogisticRegression
import nltk 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
def ProperNounExtractor(text):
    sentences = nltk.sent_tokenize(text)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        words = [word for word in words if word not in set(stopwords.words('english'))]
        tagged = nltk.pos_tag(words)
        for (word, tag) in tagged:
            if tag == 'NNP': # If the word is a proper noun
                return True
```

```python
import re
def  clean_text(df, sentence):
    df[sentence] = df[sentence].str.lower()
    df[sentence] = df[sentence].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", elem))  
    return df
df = clean_text(df,'sentence')
```

```python
from sklearn.ensemble import StackingClassifier
estimators=[('rc',RandomForestClassifier(n_estimators=6,random_state=10)),
            ('clf',ExtraTreesClassifier(n_estimators=10,criterion='gini',max_depth=13)),
            ('model',XGBClassifier(C=0.039,gamma=0.897,kernel='sigmoid'))]
c_stacking=StackingClassifier(estimators=estimators,final_estimator=LogisticRegression())
c_stacking.fit(x_train,y_train).score(x_test,y_test)
```

```python
from sklearn.ensemble import VotingClassifier
estimator=[('rc',x1),('model',x2),('clf',x3)]
elf=VotingClassifier(estimators=estimator,voting='hard')
elf.fit(x_train,y_train)
accuracy_score(y_test,elf.predict(x_test))
```
